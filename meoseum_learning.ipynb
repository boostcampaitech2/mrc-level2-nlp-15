```python
# read the json file from key and values
import json

def read_json(file):
    with open(file, 'r') as f:
        data = json.load(f)
    return data

def make_df(file, sample_csv_path):
    df_sample_csv = pd.read_csv(sample_csv_path)
    # drop context_id
    df_sample_csv = df_sample_csv.drop(columns=['context_id'])
    # rename id as context_id
    df_sample_csv = df_sample_csv.rename(columns={'id': 'context_id'})

    df = pd.DataFrame({})
    context_id = read_json(file).keys()
    answer = read_json(file).values()

    df["context_id"] = context_id
    
    column_name = file
    df[column_name] = answer
    # display(df_sample_csv)

    # merge the df_sample_csv and df
    df = pd.merge(df, df_sample_csv, on="context_id")
    
    # drop context
    df = df.drop(columns=["context"])

    return df

sample_csv = "/opt/ml/mrc-level2-nlp-15/test_retrieval_elastic_cross_ensemble_5_one_on_one.csv"
file_list = [highest, one, two, three]

df_1 = make_df(file_list[0], sample_csv)

for file in file_list:    
    
    df_2 = make_df(file, sample_csv)
    # merge df_1 and df_2 based on context_id and remove duplicates
    df_1 = pd.merge(df_1, df_2, on=["context_id", "question"])

# reorder columns of df by alphabetical order
df = df_1.copy()
df = df.reindex(sorted(df.columns, reverse=True), axis=1)
# remove the column "/opt/ml/mrc-level2-nlp-15/outputs/highest.json_x" adjacent to each other
df = df.drop(columns=["/opt/ml/mrc-level2-nlp-15/outputs/highest.json_x"])
df
```
