```python
import pandas as pd
import numpy as np
```

```python
df_top100 = pd.read_csv("/opt/ml/mrc-level2-nlp-15/test_elastic_top100.csv")
df_top100.head(2)
```

```python
# pandas settings: display full length of the column values
pd.set_option('display.max_colwidth', None)

# from the values of context and score and document_id columns leave first 5 elements only
df_top5 = df_top100.copy()
# listfiy context, score, document_id columns values
df_top5["context"] = df_top5["context"].apply(lambda x: eval(x))
df_top5["score"] = df_top5["score"].apply(lambda x: eval(x))
df_top5["document_id"] = df_top5["document_id"].apply(lambda x: eval(x))

df_top5['context'] = df_top5['context'].apply(lambda x: x[:5])
df_top5['score'] = df_top5['score'].apply(lambda x: x[:5])
df_top5['document_id'] = df_top5['document_id'].apply(lambda x: x[:5])
display(df_top5.head(2))
```

```python
# remove unnamed column
df_top5.drop(columns=['Unnamed: 0'], inplace=True)
df_top5.to_csv("/opt/ml/mrc-level2-nlp-15/test_elastic_top5.csv", index=False)
```

```python
# read from json and convert into dataframe
series = pd.read_json('./best_predictions.json', typ='series')
df = pd.DataFrame({'index':series.index, 'answer':series.values})
df
```

```python
df.tail()
```

```python
# # remove "" from the dataframe value
df = df.applymap(lambda x: x.replace('\"', ''))
# remove backslash \ from the dataframe value
df = df.applymap(lambda x: x.replace('\\', ''))

# # remove " from the dataframe value
# df = df.applymap(lambda x: x.replace('\"', ''))
df = df.applymap(lambda x: x.replace("\'", ""))

# # remove '' from the dataframe value
df = df.applymap(lambda x: x.replace("’", ""))
df = df.applymap(lambda x: x.replace("‘", ""))

# # remove ""

df.tail()
```

```python
import json
from easydict import EasyDict

json_file = EasyDict()
answers = df.to_dict(orient='records')
for item in answers:
    json_file[item['index']] = item['answer']
# save json file as csv
json_file
```

```python
import pandas as pd
# save json format as csv format
df = pd.DataFrame(json_file)
df.to_csv('./best_predictions_fixed.csv', index=False)
```
