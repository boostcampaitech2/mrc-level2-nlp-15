```python
k = 0
c = 0
trains = datas['train']
from tqdm import tqdm
for i in tqdm(range(len(trains))):
    query= trains[i]['question']
    try:
        res = es.search(index=INDEX_NAME, q=query, size=3)
        x = res['hits']['hits']
        for docu in x:
            if docu['_source']['text'] == trains[i]['context']:
                c+=1
                break
        k+=1
    except:
        print(i)
        pass
print(c/k)
print(k)
```
