```python
import json
```

```python
x = []
for idx,i in enumerate(range(1,6)):
    if idx == 3:
        continue
    with open(f'output ({i}).json') as x1:
         x.append(json.load(x1))
```

```python
from collections import defaultdict,OrderedDict
answers = defaultdict(OrderedDict)
for json_data in x:
    for id,j in json_data.items():
        try:
            answers[id][j] += 1
        except:
            answers[id][j] = 1
```

```python
answers
```

```python
result = {}
for i,j in answers.items():
    result[i] = list(j)[0]
```

```python
a = json.dump(result)
# json_val = json.dumps(result)
```

```python
with open('submission.json', 'w') as outfile:
    json.dump(result, outfile)
```
